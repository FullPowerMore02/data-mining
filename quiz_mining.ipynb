```python
#3.1นับจำนวนแต่ละประเทศในคอลัมน์ 'native-country'
country_counts = df[' native-country'].value_counts()

# แสดงตัวอย่างประเทศที่ปรากฏในข้อมูล
print(country_counts.head())

# หาจำนวนคนที่มีถิ่นกำเนิดเป็น 'Thailand'
thai_count = df[df[' native-country'] == 18].shape[0]
print("จำนวนคนไทย:", thai_count)
```

```python
#3.2สร้างตัวแปร capital-total = capital-gain - capital-loss
df['capital-total'] = df[' capital-gain'] - df[' capital-loss']
df
```

```python
#3.3ตัวอย่างที่ถิ่นกำเนิดใดมีค่าเฉลี่ยของ capital-total มากที่สุด
# หาค่าเฉลี่ยของ capital-total ในแต่ละกลุ่ม (ตาม native-country)
mean_capital_total_by_country = df.groupby(' native-country')['capital-total'].mean()

# หาค่าเฉลี่ยของ capital-total ที่มากที่สุด
max_mean_capital_total = mean_capital_total_by_country.max()

# หาตัวอย่างที่มีค่าเฉลี่ยของ capital-total เท่ากับค่าที่มากที่สุด
country_with_max_mean_capital_total = mean_capital_total_by_country[mean_capital_total_by_country == max_mean_capital_total].index[0]

print("ตัวอย่างที่ถิ่นกำเนิด '{}' นั้นคือ India มีค่าเฉลี่ยของ capital-total มากที่สุด: {}".format(country_with_max_mean_capital_total, max_mean_capital_total))
```

```python
#3.4 อาชีพใด (occupation) มีจำนวนตัวอย่างที่มีรายได้ >50K มากที่สุด
# หาจำนวนตัวอย่างที่มีรายได้ >50K ในแต่ละกลุ่ม (ตาม occupation)
count_high_income_by_occupation = df[df[' income'] == 1].groupby(' occupation').size()

# หาจำนวนตัวอย่างที่มีรายได้ >50K มากที่สุด
max_count_high_income = count_high_income_by_occupation.max()

# หาอาชีพที่มีจำนวนตัวอย่างที่มีรายได้ >50K มากที่สุด
occupation_with_max_count_high_income = count_high_income_by_occupation[count_high_income_by_occupation == max_count_high_income].index[0]

print("อาชีพ Exec-managerial'{}' มีจำนวนตัวอย่างที่มีรายได้ >50K มากที่สุด: {}".format(occupation_with_max_count_high_income, max_count_high_income))
```
